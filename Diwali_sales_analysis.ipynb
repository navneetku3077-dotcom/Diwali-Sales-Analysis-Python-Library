```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import csv file
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
# to avoid encoding error use,'unicode_escape'
```

```python
df.shape
print(df.shape)
df.head()
```

```python
df.info()
```

```python
# removing unrelated /blank column
df.drop(['Status', 'unnamed1'],axis=1,inplace=True)
df.shape
```

```python
#check null value
pd.isnull(df).sum()
```

```python
#drop null value
df.dropna(inplace=True)
df.info() #null value drop successful
```

```python
df['Amount']=df['Amount'].astype('int') #change the data type of amount
df['Amount'].dtypes #change successfully
```

```python
df.columns #name of column
```

```python
df.rename(columns={'Martial_Status':'shadi'})
```

```python
#describe() method return description of the data in the dataframe(i.e.count,mean,std,etc)
df.describe()
```

```python
#use describe() for specific column
df[['Age','Orders','Amount']].describe()
```

```python
#exploratory data analysis
# plotting a bar chart for gender vs total amount
plt.figure(figsize=(6,4))
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

**From this graph we can see that most of the buyers are female and even purchasing power of female are greater than men**

### AGE

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
#hue use for legend
for bars in ax.containers: #for loop are used to label on top of bars
    ax.bar_label(bars)
```

```python
#TOtal Amount vs age group
plt.figure(figsize=(6,4))
Sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=Sales_age,palette='viridis',hue='Age Group') #Tip use palette when using hue
```

**From above graphs we can see that most of the buyers are of age group between 26-35 yrs female**

##  State

```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='State',y='Orders',data=sales_state,color='purple')

for i in ax.containers:
    ax.bar_label(i)
```

```python
#Total amount from top 10 states
plt.figure(figsize=(15,6))
state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(x='State',y='Amount',data=state_amount,color='blue')
labels=[f'${x/1000000:.2f}M'for x in state_amount['Amount']] #adding label bars in respect to amount
for i in ax.containers:
    ax.bar_label(i,labels=labels)
```

**From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively**

## Marital Status

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(data=df,x='Marital_Status',palette=['red','black'],hue='Marital_Status')
for x in ax.containers:
    ax.bar_label(x)
ax.legend(title='Marital_status',labels=['Married','Unmarried'])
```

```python
Marital_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=Marital_status)
```

**From above graph we can see that most of the buyers are married women and they have high purchasing power**

## Occupation ##

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation',color='green',legend=False)
for x in ax.containers:
    ax.bar_label(x)
```

```python
occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=occupation,color='yellow')
```

**From above we can see that most of the buyers are working in IT,aviation and Healthcare sector**

## Product Category ##

```python
plt.figure(figsize=(30,8))
ax=sns.countplot(data=df,x='Product_Category')
for x in ax.containers:
    ax.bar_label(x)
```

```python
plt.figure(figsize=(20,6))
product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=product,color='red')
```

**From the above graph we can see that the max revenue recieved from Food ,clothing & apparel and electronic & gadget**

```python
product_ID=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=product_ID,x='Product_ID',y='Orders',color='brown')
```

## Conclusion: ##

**Married women age group 26-35yrs from UP,Maharastra and karnataka working in IT,Healthcare and Aviation are more likely to buy product from Food, Clothing and electronic category**
